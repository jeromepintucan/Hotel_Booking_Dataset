```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
hotel = pd.read_excel('./DATAS//HotelFinalDataset.xlsx')
```

```python
hotel.head()
```

```python
hotel.drop(columns=hotel.columns[0], inplace=True)
```

```python
hotel.isnull().sum()
```

```python
hotel.loc[hotel['State'].isnull()]
```

```python
hotel[hotel['City'] == 'Bemelen'].fillna(value='Limburg', axis=1, inplace=True)
```

```python
hotel[25:50] = hotel[25:50].fillna('Lumburg')
```

```python
hotel.loc[hotel['State'].isnull()]
```

```python
hotel[(hotel['City'] == 'Breda') & (hotel['State'].isnull())] = hotel[(hotel['City'] == 'Breda') & (hotel['State'].isnull())].replace(np.nan, 'North Brabant')
```

```python
hotel[(hotel['City'] == 'Breda')]
```

```python
hotel.head()
```

```python
hotel.loc[(hotel['City'] == 'Delft') & (hotel['State'].isnull())] = hotel.loc[(hotel['City'] == 'Delft') & (hotel['State'].isnull())].replace(np.nan, 'South Holland')
```

```python
hotel.isnull().sum()
```

```python
hotel.loc[hotel['State'].isnull()]
```

### FIND THE UNIQUE VALUE IN THE CITY COLUMN

```python
city_list = []

for i in hotel.loc[hotel['State'].isnull()]['City'].values:
    if i not in city_list:
        city_list.append(i)

print(city_list)
```

### CREATE THE COUNTERPART STATE OF EACH CITY

```python
state_list = ['North Brabant', 'North Braband', 'Gronigen', 'North Holland', 'South Holland', 'North Holland', 'Friesland', 'Limburg', 'Mpumalanga', 'Gelderland', 'Limburg', 'The Hague', 'Central Netherland', 'Limburg', 'Zeeland', 'Gelderland', 'North Holland', 'North Holland']
```

### CONVERT TWO (2) LIST INTO A DICTIONARY


```python
city_state = {}

for city in range(len(city_list)):
    city_state[city_list[city]] = state_list[city]

print(city_state)
```

### FILL THE NAN VALUES WITH THE CORRESPONDING VALUE IN THE CITY COLUMN

```python
hotel.loc[hotel['State'].isnull(), 'State'] = hotel.loc[hotel['State'].isnull(), 'State'].fillna(hotel['City'].map(city_state))
```

### MAKE SURE THE STATE COLUMN HAS NO NAN VALUES

```python
hotel.isnull().sum()
```

```python
hotel.loc[hotel['ReviewsCount'].isnull()]
```

```python
hotel['ReviewsCount'].fillna(0, inplace=True)
```

```python
hotel.isnull().sum()
```

```python
hotel['Rating'].fillna(0, inplace=True)
```

```python
hotel.isnull().sum()
```

```python
hotel.head()
```

```python
hotel.dtypes
```

```python
hotel['Price'] = hotel['Price'].replace('₹', '')
```

```python
hotel['Price'] = hotel['Price'].str.replace('₹', '')
```

```python
hotel['Price'] = hotel['Price'].str.replace(',', '')
```

```python
hotel['Price'] = hotel['Price'].str.replace(' ', '')
```

```python
hotel.head()
```

```python
hotel['Price'] = hotel['Price'].astype(int)
```

```python
hotel.dtypes
```

### START TO ANALYZE THE DATA

```python
hotel.head()
```

```python
hotel.describe()
```

```python
hotel['Price (USD)'] = round(hotel['Price'] * 0.013, 2)
```

```python
hotel.head()
```

```python
type_price = hotel.groupby(['Type'], as_index=False).agg({'Price (USD)': 'mean'}).sort_values(by='Price (USD)', ascending=False)[0:20]
type_price
```

```python
hotel.groupby(['Type'], as_index=False).agg({'Price (USD)': 'sum'}).sort_values(by='Price (USD)', ascending=True)
```

### Most Expensive Hotel Booking Cost Per Room Type

```python
sns.set()

fig = plt.figure(figsize=(16,7))
sns.barplot(y='Type', x='Price (USD)', data=type_price, palette='pastel')
plt.title('Hotel Booking Cost Per Type of Room', size=12)

plt.show()
```

```python
hotel.head()
```

### STATES WITH THE HIGHEST NUMBER OF HOTEL

```python
hotel_city_count = hotel.groupby(['State'], as_index=False).agg({'Name': 'count'}).sort_values(by='State', ascending=True)
hotel_city_count
```

```python
sns.set()
fig = plt.figure(figsize=(16,6))

z = range(0, 100, 5)

sns.barplot(y='State', x='Name', data=hotel_city_count)
plt.ylabel('States')
plt.xlabel('Number of Hotels')
plt.xticks(z)
plt.title('STATES WITH THE HIGHEST NUMBER OF HOTEL')

plt.show()
```

### STATES WITH THE MOST EXPENSIVE HOTELS

```python
hotel.head()
```

```python
hotel_state_price = hotel.groupby(['State'], as_index=False).agg({'Price (USD)': 'mean'}).sort_values(by='State', ascending=True)
hotel_state_price
```

```python
fig = plt.figure(figsize=(16,6))

sns.barplot(y='State', x='Price (USD)', data=hotel_state_price.sort_values(by='Price (USD)', ascending=False))
plt.title('STATES WITH THE MOST EXPENSIVE HOTEL')

plt.show()


# The graph shows that Maastricht, Lumburg and Amsterdam have the most expensive hotels
```
